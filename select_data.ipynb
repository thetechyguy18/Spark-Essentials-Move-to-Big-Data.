## Different ways to select the data 

## creating a spark session

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
        .master("local[*]") \
        .appName("SparkTest") \
        .getOrCreate()

spark
```

## creating a dataframe by reading a data file 

```python
sales = spark.read.format('csv'). \
     option('InferSchema',True). \
     option('Header',True). \
     load("D:\Datasets\SQL - Retail Sales Analysis_utf .csv")
```

```python
sales.printSchema()
```

### 1. select() (MOST COMMON)

```python
sales.select('customer_id').show(3)
```

```python
sales.select('*').show(2)
```

### 2. col() (FUNCTION STYLE)

```python
from pyspark.sql.functions import *
sales.select(column('transactions_id'),col('sale_date') ,col('gender'),(col('quantiy') + 5).alias('increased_price')).show(4)
```

### 3. selectExpr() (SQL-LIKE üî•)

```python
sales.selectExpr('transactions_id',
               'quantiy * 5 as better_qnty').show(5)
```

### 4. DataFrame Column Object (df["col"])

```python
sales.select(["transactions_id"]).show(5)
```

```python
sales.select(["transactions_id","quantiy"]).show(5)
```
