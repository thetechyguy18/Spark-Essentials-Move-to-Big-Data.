##WINDOW FUNCTIONS/TRANSFORMATION 

### RANK, DENSE RANK, ROW_NUMBER

```python
df = spark.read.option('inferschema',True).option('header',True).csv('/Volumes/learn_spark/bronze/datasets/Datasets/student_marks_200_rows.csv')
```

```python
df.show()
```

## ROW_NUMBER 

### this treats every rank or marks as different whether it is same or not

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *


window_spec = Window.partitionBy('subject').orderBy(desc('marks'))

df.withColumn('row_number', row_number().over(window_spec)).display()

```

## Rank 
### It gives the unique rank to same marks but the afterward student will get a rank bewlow that 

```python
df.withColumn('rank', rank().over(window_spec)).display()
```

## Dense Rank
### Every Students who got same marks treats as same also the next student got the rank after that without skipping the rank

```python
df.withColumn('rank', dense_rank().over(window_spec)).display()
```

```python
df.withColumn('row_number', row_number().over(window_spec))\
    .withColumn('rank', rank().over(window_spec))\
        .withColumn('desce_rank', dense_rank().over(window_spec)).display()
```
