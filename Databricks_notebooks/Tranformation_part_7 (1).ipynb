# Tranformation part 7
- ### - GroupBy, agg, sum, avg, collection_list, when otherwise, pivot


```python
df = spark.read.option('header',True).option('inferschema',True).csv('/Volumes/learn_spark/bronze/datasets/Datasets/SalesData_Large.csv')
```

```python
df.show()
```

### GROUPBY 

```python
from pyspark.sql.functions import *;
df.groupBy('Region').agg(sum('SalesAmount').alias('Total_sales'), avg("SalesAmount").alias('avg_sales'),round(sum('Profit'),2).alias('profit')).display()
```

### Collection list

```python
df.groupBy('Region').agg(collect_list('Product')).display()
```

### PIVOT

```python
df.groupBy('Region').pivot('Product').agg(sum('SalesAmount')).display()
```

### WHEN AND OTHERWISE

```python
df.withColumn('good_product',when(col('Quantity') > 10,'yes').otherwise('No')).display()
```

```python
df.withColumn('good_product',when((col('Quantity') > 10) | (col('Profit') > 10000),'yes').otherwise('No')).display()
```
