## DATE FUNCTIONS WITH ITS PRACTICAL 

### DATE_FORMAT, HOW TO ACCESS MONTH, CURR_DATE, DATE_ADD, DATE_SUB, DATE_DIFF 

```python
employee = spark.read.format('csv')\
    .option('Header',True)\
            .load('/Volumes/learn_spark/bronze/datasets/employee_data_with_nulls_duplicates.csv')
```

```python
employee.printSchema()
```

```python
emp = employee.withColumn('join_date', date_format('join_date','YYYY-MM-DD'))
```

```python
emp.printSchema()
```

```python
employee.limit(20).display()
```

```python
from pyspark.sql.functions import *
employee.withColumn('curr_date',current_date()).display()
```

### calculated yrs of experience of an emp 

```python
employee.withColumn('curr_date',current_date())\
    .withColumn('yrs_of_exp', (col('join_date')-col('curr_date'))/365).display()
```

### Adding a date to find the promotion date 

```python
employee.withColumn('curr_date',current_date())\
    .withColumn('promotion_date', date_add('curr_date',180)).show()
```

### findiing the diff b/w dates 

```python
employee.withColumn('curr_date',current_date())\
    .withColumn('promotion_date', date_add('curr_date',180))\
        .withColumn('Days_rem_for_promotion', datediff('promotion_date','curr_date')).show(20)
```

### going back to your previous date

```python
employee.withColumn('month', date_format(col('join_date'),'M')).display()
```
