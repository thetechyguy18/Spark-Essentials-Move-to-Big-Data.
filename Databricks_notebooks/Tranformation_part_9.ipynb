## Joins 
### inner, left, right, full, left-anti, left-semi

#### Data Ingestion

```python
emp_data = [
    (1, "Alice", 10, 60000),
    (2, "Bob", 20, 70000),
    (3, "Charlie", 10, 80000),
    (4, "David", 30, 75000),
    (5, "Eva", None, 65000),
    (6, "Frank", 40, 72000)
]

emp_cols = ["emp_id", "emp_name", "dept_id", "salary"]

emp_df = spark.createDataFrame(emp_data, emp_cols)
emp_df.show()



dept_data = [
    (10, "Engineering"),
    (20, "HR"),
    (30, "Finance"),
    (50, "Marketing")
]

dept_cols = ["dept_id", "dept_name"]

dept_df = spark.createDataFrame(dept_data, dept_cols)
dept_df.show()
```

## Inner Join
### Basically shares the common data of both the tables

```python
emp_df.join(dept_df,'dept_id', 'inner').display()
```

## LEFT JOIN
### It shows all the data from your left/first table and not matching i.e Ids not in 2nd table as null

```python
emp_df.join(dept_df,'dept_id', 'left').display()
```

## RIGHT JOIN
### It shows all the data from your right/second table and not matching i.e Ids not in 1st table as null

```python
emp_df.join(dept_df,'dept_id', 'right').display()
```

## full join
### gives all the data from the two tables without worring about null keys (combination of both i.e left + right)

```python
emp_df.join(dept_df,'dept_id', 'full').display()
```

## left Anti
### Very Imp gives result which is not in left join as name suggest anti

```python
emp_df.join(dept_df,'dept_id', 'left').display()
emp_df.join(dept_df,'dept_id', 'leftanti').display()
```

## Left semi 
### show only left table data + only distinct ids

```python
emp_df.join(dept_df,'dept_id', 'left').display()
emp_df.join(dept_df,'dept_id', 'left_semi').display()
```
