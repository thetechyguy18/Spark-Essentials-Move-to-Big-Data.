## FINDING NULL DATA  & filtering it based on some columns 


```python
from pyspark.sql.functions import * 
employee_not_null = employee.filter(col('employee_id').isNotNull())
```

```python
employee_not_null.count()
```

```python
employee_unique = employee_not_null.distinct()
```

## fiding the null columns 

```python
employee_unique.display()
```

## FInding records in a specifc creteria


```python
employee_unique.filter(col('city').isin('Bangalore','Pune')).display()
```

## Data Cleaning

```python
complyee_not_blank= employee.dropna('all')
```

```python
complyee_not_blank.display()
```

```python
complyee_cleaned = complyee_not_blank.dropna(subset = ['city','salary'])
```
