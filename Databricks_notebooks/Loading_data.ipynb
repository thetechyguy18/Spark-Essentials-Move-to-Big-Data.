### CAST TRANSFORMATION USE TO CHANGE THE DATATYPE OF EXISTING COLUMN

```python
from pyspark.sql.functions import *
students_df_new = students.select('*')\
    .withColumn('student_id', col('student_id').cast('int'))\
    .withColumn('age', col('age').cast('int'))\
    .withColumn('marks', col('marks').cast('int'))
```

```python
students_df_new.printSchema()
```

```python
students_df_new.show(10)
```

### FILTER OR WHERE USED TO FILTER THE RECORDS

```python
students_df_new.filter(column('marks') > 80).count()
```

```python
students_df_new.where(column('gender') == 'F').count()
```

###  Alias used to change the name of an existing column or for a new column after col('column).alias('new_column_name')


### withColumnRenamed used to change that paricular name



```python
students_df_new.select('*')\
    .withColumnRenamed('age', 'new_age').show(10)
```
