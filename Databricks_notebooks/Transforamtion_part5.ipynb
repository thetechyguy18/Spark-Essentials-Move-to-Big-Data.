```python
df = spark.read.format('csv').option('Header',True).load('/Volumes/learn_spark/bronze/datasets/employee_data_with_nulls_duplicates.csv')
```

### FIll na operation

```python
df.fillna("not available", subset= ["city"]).show(20)
```

### union BY NAME

```python
data1 = [
    (1, "Amit", "IT"),
    (2, "Neha", "HR"),
    (3, "Rahul", "Finance")
]

df1 = spark.createDataFrame(
    data1,
    ["emp_id", "name", "department"]
)
```

```python
data2 = [
    ("Sales", 4, "Priya"),
    ("IT", 5, "Karan")
]

df2 = spark.createDataFrame(
    data2,
    ["department", "emp_id", "name"]
)
```

```python
df1.unionByName(df2).show()
```

```python
data3 = [
    ( 4, "Priya","Sales"),
    ( 5, "Karan","IT")
]

df3 = spark.createDataFrame(
    data3,
    [ "emp_id", "name","department"]
)
```

### UNION 

```python
df1.union(df3).show()
```

### lower


```python
from pyspark.sql.functions import *
df.select(lower('name').alias('new_name'), "employee_id","department",'salary','city','join_date').show(10)
```

### UPPER

```python
df.select(upper('name')).show(10)
```

### init cap

```python
df.select(initcap('lower(name)')).show(10)
```
